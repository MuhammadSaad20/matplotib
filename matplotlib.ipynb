Barchart

```python
company=["GOOGLE","AMAZON","MSFT","FACEBOOK"]
revenue=[90,136,89,27]
profit=[40,2,32,12]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.xticks(ypos,company)
plt.ylabel("Revenue ($Bin)")
plt.title("US Tech Stock")
plt.xlabel("Companies")
plt.bar(ypos,revenue,label="Revenue")
plt.bar(ypos,profit,label="Profit")

plt.legend()
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(ypos,company)
plt.ylabel("Revenue ($Bin)")
plt.title("US Tech Stock")
plt.xlabel("Companies")
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")

plt.legend()
```

```python
plt.yticks(ypos,company)
plt.xlabel("Revenue ($Bin)")
plt.title("US Tech Stock")
plt.ylabel("Companies")
plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")

plt.legend()
```

Historogram

```python
blood_sugar=[113,85,90,150,159,88,92,115,125,80,77,82,120]

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color="green")

```

```python
blood_sugar_man=[113,85,90,150,159,88,92,115,125,80,77,82,120]
blood_sugar_woman=[65,98,50,130,139,78,52,111,125,90,79,84,130]
plt.xlabel('Sugar range')
plt.ylabel('Num of Patient')
plt.title('Blood Sugar Analysis')


#bins are distributin among historogarm
plt.hist([blood_sugar_man,blood_sugar_woman],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['man','woman'])
plt.legend()

```

```python
blood_sugar_man=[113,85,90,150,159,88,92,115,125,80,77,82,120]
blood_sugar_woman=[65,98,50,130,139,78,52,111,125,90,79,84,130]
plt.ylabel('Sugar range')
plt.xlabel('Num of Patient')
plt.title('Blood Sugar Analysis')


#bins are distributin among historogarm
plt.hist([blood_sugar_man,blood_sugar_woman],bins=[80,100,125,150],rwidth=0.95,color=['green','orange'],label=['man','woman'],orientation='horizontal')
plt.legend()

```

Piechart

```python
exp_val=[1400,600,300,420,240]
exp_label=["Home Rent","Food Item","Phone Bill","Car","Other Utilize"]
```

```python
plt.axis("equal")
plt.pie(exp_val,labels=exp_label,radius=1.5,autopct="%0.1f%%",shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_val,labels=exp_label,radius=1.5,autopct="%0.1f%%",shadow=True,explode=[0,0.1,0.1,0,0],startangle=180)
plt.show()
```
